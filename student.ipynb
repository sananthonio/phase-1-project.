```python
#dropping the initial columns that are not of any significance.
columns_to_drop = [
    'Airport.Code',
    'Airport.Name',
    'Schedule',
    'Air.carrier',
    'Latitude',
    'Longitude',  
    'FAR.Description']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
#finding percentage of remaining missing value.
df.isna().mean()*100
```

```python
#describing my data.
df.describe()
```

```python
df.dropna(subset=['Location', 'Country', 'Injury.Severity', 'Aircraft.damage', 'Registration.Number', 'Make', 'Model', 'Amateur.Built', 'Number.of.Engines', 'Engine.Type', 'Purpose.of.flight', 'Report.Status',], inplace=True)
```

```python

df['Total Injuries'] = df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries'] + df['Total.Minor.Injuries']
```

```python
#loading cleaned data
df.to_csv('cleaned_Aviation_data.csv', index=False)
```

```python
#ACCIDENTS BASED ON PURPOSE OF FLIGHT.
fig,ax = plt.subplots(figsize = (15,10))
ax.bar(df['Purpose.of.flight'],df['Total Injuries'])
fig.set_tight_layout(True)
ax.set_xlabel('Purpose.of.flight')
ax.set_ylabel('Number of accidents and incidents')
ax.set_title('Top 10 most affected sectors in the industry')
plt.show();
```
